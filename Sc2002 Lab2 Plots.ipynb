```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('results2.csv')

# Sort data by 'Number of Vertices' and ' Number of Edges'
df_sorted_vertices = df.sort_values(by='Number of Vertices')
df_sorted_edges = df.sort_values(by=' Number of Edges')

# Extract columns for sorted data
vertices = df_sorted_vertices['Number of Vertices']
edges = df_sorted_edges[' Number of Edges']
matrix_array_pq_vertices = df_sorted_vertices[' Matrix & Array PQ(ns)']
list_min_heap_pq_vertices = df_sorted_vertices[' List & Minimizing Heap PQ(ns)']
matrix_array_pq_edges = df_sorted_edges[' Matrix & Array PQ(ns)']
list_min_heap_pq_edges = df_sorted_edges[' List & Minimizing Heap PQ(ns)']

# Plot for Number of Vertices vs PQ methods
plt.figure(figsize=(10,5))
plt.plot(vertices, matrix_array_pq_vertices, label="Matrix & Array PQ", marker='o')
plt.plot(vertices, list_min_heap_pq_vertices, label="List & Minimizing Heap PQ", marker='x')
plt.title('Comparison of PQ Methods vs Number of Vertices')
plt.xlabel('Number of Vertices')
plt.ylabel('Time (ns)')
plt.legend()
plt.grid(True)
plt.ticklabel_format(style='plain', axis='x')  # Disable scientific notation for x-axis
plt.show()

# Plot for Number of Edges vs PQ methods
plt.figure(figsize=(10,5))
plt.plot(edges, matrix_array_pq_edges, label="Matrix & Array PQ", marker='o')
plt.plot(edges, list_min_heap_pq_edges, label="List & Minimizing Heap PQ", marker='x')
plt.title('Comparison of PQ Methods vs Number of Edges')
plt.xlabel('Number of Edges')
plt.ylabel('Time (ns)')
plt.legend()
plt.grid(True)
plt.ticklabel_format(style='plain', axis='x')  # Disable scientific notation for x-axis
plt.show()

# Filter data for a fixed number of vertices (for example, vertices = 1000)
fixed_vertices = 30
df_fixed_vertices = df[df['Number of Vertices'] == fixed_vertices]

# Box plot for Matrix & Array PQ and List & Minimizing Heap PQ based on different number of edges
plt.figure(figsize=(10, 6))

# Create the box plot comparing the time for the two PQ methods at different edge counts
df_fixed_vertices.boxplot(column=[' Matrix & Array PQ(ns)', ' List & Minimizing Heap PQ(ns)'],
                          by=' Number of Edges', grid=True)



plt.xlabel('Number of Edges')
plt.ylabel('Time (ns)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
